# Iteration Order(Cstyle & f style)

```python
print(b)
print(np.char.center("=",50,"="))
for x in np.nditer(b, order="C"):
    print(x)
print(np.char.center("=",50,"="))
for x in np.nditer(b,order="F"):
    print(x)
```

# Joining Arrays

```python
a = np.array([[2,3],[4,5]])
print("first array \n ", a)
b = np.array([[6,7],[8,9]])
print("second array \n", b)
print("\n Joining two arrays along 0 axis\n")
print(np.concatenate((a,b)))
print("\n Joining two arrays along 1 axis\n")
print(np.concatenate((a,b),axis=1))
```

# Resizing Array

```python
a = np.array([[1,2,3],[4,5,6]])
print(np.char.center("=",50,"="))
print(a)
print(np.char.center("=",50,"="))
print(a.shape)
print("\n")
b = np.resize(a,(3,2))
print(b)
print(np.char.center("=",50,"="))
print(b.shape)
c = np.resize(a,(3,3))
print(c)
print(np.char.center("=",50,"="))
print(c.shape)
```

### What is Pandas

Pandas is a tool for data processing which helps in data Analysis.
It provides functions and methods to efficiently manipulate large datasets.

### Series

**`Data Structures in Pandas`**

- `Series` (One dimensional Array)
- `DataFrame` (two-dimensional Array)

`Series` One-dimensional arrays with labels. It can contain any datatype including integers,string, floats, Python Objects and more.

`DataFrame`Two-dimensional data structures with labels. We use lables to locate data.

```python
import pandas as pd
print(pd.__version__)
```

### Series create, manipulate, query, delete

```python
# Creating a series from a list
arr = [0,1,2,3,4 ]
s1 = pd.Series(arr)
s1
```

```python
order = [1,2,3,4,5]
s2 = pd.Series(arr,index=order)
s2
```

```python
n = np.random.randn(5)
index = ["a","b","c","d","e"]
s2 = pd.Series(n,index=index)
s2
```

```python
# Create A series from a dictioanry
d = {"a":1, "b":2,"c":3,"d":4,"e":5}
s3 = pd.Series(d)
s3
```

```python
# Modify Index of the series
print(s1)
```

```python
s1.index =["A","B","C","D","E"]
s1
```

```python
# slicing
s1[:3]
```

```python
s1[:-2]
```

```python
# Appending
s4 = pd.concat([s1,s3])
s4
```

```python
s4.drop("e")
```
